```python
import pandas as pd
```

```python
data = pd.read_csv(r'E:\WorldBank-Data-Analysis-Project\DATA_AFTER_PREPROCESSING\dataset_merged.csv')
data
```

```python
data = data[data['Year'] != 2024]
```

```python
# Lấy danh sách các cột cần tìm outliers (loại trừ Country và Year)
features = ['Agri_GDP', 'Credit_Private', 'Elec_Access', 'Emp_Services', 
            'FDI_GDP', 'GDPC_2015', 'Indus_GDP', 'Internet_Usage', 'Serv_GDP']

for col in features:
    # Tính toán Q1, Q3 và khoảng cách IQR
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Xác định giới hạn trên và dưới
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Lọc các dòng chứa outliers
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    # In kết quả
    print(f"{col} :")
    if not outliers.empty:
        for _, row in outliers.iterrows():
            print(f"  Country: {row['Country']}, Year: {row['Year']}, Value: {row[col]}")
    else:
        print("  Không có outliers")
    print("-" * 40)
```

```python
import pandas as pd
# Import class DataProcessor từ file preprocess.py
from processing import DataProcessor

# 1. Khởi tạo đối tượng xử lý dữ liệu với dữ liệu đầu vào là data_copy
processor = DataProcessor(data_copy)

# 2. Định nghĩa danh sách các cột bạn muốn xử lý khuyết (hoặc để None để tự quét toàn bộ cột số)
features_to_impute = [
    'Agri_GDP', 'Elec_Access', 
    'FDI_GDP', 'GDPC_2015', 'Indus_GDP', 'Internet_Usage', 'Serv_GDP'
]

# 3. Gọi hàm xử lý missing value theo phương pháp multivariate (IterativeImputer)
# Hàm này sẽ tự động gọi _impute_multivariate ở bên trong
data_copy_imputed = processor.handle_missing_values(
    method='multivariate',
    tree_model='random_forest',
    features=features_to_impute,
    drop_threshold=0.5 # Bỏ qua các cột có tỉ lệ khuyết > 50% nếu có
)

# Giờ data_copy_imputed đã là một DataFrame mới đã được điền các giá trị khuyết
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_lower_correlation(data, filename, title):
    # Tạo figure
    plt.figure(figsize=(10, 8))
    
    # Tính ma trận tương quan
    corr = data.corr()
    
    # Tạo mask để che đi nửa trên của ma trận (upper triangle)
    mask = np.triu(np.ones_like(corr, dtype=bool))
    
    # Vẽ heatmap kết hợp với mask
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', 
                linewidths=0.5, square=True, mask=mask)
    
    plt.title(title)
    plt.tight_layout()
    
    # Lưu biểu đồ dưới dạng PDF
    plt.savefig(filename, format='pdf')
    plt.close()

# 1. Vẽ và lưu biểu đồ cho data_copy 
plot_lower_correlation(
    data_copy.drop(columns=['Country', 'Year'], errors='ignore'), 
    'correlation_data_copy.pdf', 
    'Correlation Matrix - Original Data'
)

# 2. Vẽ và lưu biểu đồ cho data_copy_imputed 
plot_lower_correlation(
    data_copy_imputed.drop(columns=['Country', 'Year'], errors='ignore'), 
    'correlation_data_copy_imputed.pdf', 
    'Correlation Matrix - Imputed Data'
)
```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

def plot_kde_comparison(df_before, df_after, filename='kde_comparison.pdf'):
    features = df_before.columns
    n_features = len(features)
    
    # Thiết lập layout: tối đa 4 biểu đồ trên 1 hàng ngang để chiếm hết chiều rộng trang
    cols_per_row = 4
    n_rows = math.ceil(n_features / cols_per_row)
    
    # Tạo figure với chiều ngang lớn (20)
    fig, axes = plt.subplots(n_rows, cols_per_row, figsize=(20, 4 * n_rows))
    axes = axes.flatten() # Làm phẳng mảng axes để dễ lặp
    
    for i, col in enumerate(features):
        # Vẽ KDE đè lên nhau (không có histogram)
        sns.kdeplot(data=df_before[col], ax=axes[i], label='Before', color='blue', linewidth=2)
        sns.kdeplot(data=df_after[col], ax=axes[i], label='After', color='red', linewidth=2, linestyle='--')
        
        axes[i].set_title(f'Feature: {col}')
        axes[i].legend()
        
    # Xóa các subplots dư thừa nếu số features không chia hết cho 4
    for j in range(n_features, len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    plt.savefig(filename, format='pdf', bbox_inches='tight')
    plt.close()

# Thực thi hàm
plot_kde_comparison(data_copy.drop(columns=['Country', 'Year'], errors='ignore'), data_copy_imputed.drop(columns=['Country', 'Year'], errors='ignore'))
```
